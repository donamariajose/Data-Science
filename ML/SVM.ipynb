```python
# data model implimentation using svm algorithm
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

# load data
data=pd.read_csv('C:/Users/donad/OneDrive/Desktop/Internship/1_Machine Learning/4_Supervised_Learning/Classification/Datasets/Fish.csv')
data
```

```python
# check for null value
data.isnull().sum()
```

```python
# check for dupplicates
data.duplicated().sum()
```

```python
# splitting to train and test
train,test=train_test_split(data,test_size=0.1)
train
```

```python
test
```

```python
trainx=train.iloc[:,1:]
trainy=train.iloc[:,0]
testx=test.iloc[:,1:]
testy=test.iloc[:,0]

# model building
# linear
model_linear=SVC(kernel='linear')
predlinear=model_linear.fit(trainx,trainy).predict(testx)
acc=np.mean(predlinear==testy)
acc
```

```python
# ploynomial
model_poly=SVC(kernel='poly')
predpoly=model_poly.fit(trainx,trainy).predict(testx)
acc1=np.mean(predpoly==testy)
acc1
```

```python
# gaussian-radian basis fn
model_rbf=SVC(kernel='rbf')
predrbf=model_rbf.fit(trainx,trainy).predict(testx)
acc2=np.mean(predrbf==testy)
acc2
```

```python
# sigmoid
model_sigmoid=SVC(kernel='sigmoid')
predsig=model_sigmoid.fit(trainx,trainy).predict(testx)
acc3=np.mean(predsig==testy)
acc3
```
