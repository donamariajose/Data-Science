```python
# MACHINE LEARNING PROJECT
# Flight delay prediction
# Using collected data from the past, predict the hours/minutes/seconds each flight gets delayed.

# import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.preprocessing import LabelEncoder
from sklearn import tree
```

```python
# import dataset 1 and display it
Jan2019 = pd.read_csv(r"C:\Users\donad\OneDrive\Desktop\Internship\1_Machine Learning\project_Datasets\Flight_delay_prediction\Jan_2019_ontime.csv")
Jan2019
```

```python
#import dataset 2 and display it
Jan2020 = pd.read_csv(r"C:\Users\donad\OneDrive\Desktop\Internship\1_Machine Learning\project_Datasets\Flight_delay_prediction\Jan_2020_ontime.csv")
Jan2020
```

```python
# Combining 2 datasets
flight_del_pred_df = pd.concat([Jan2019, Jan2020], ignore_index=True)
flight_del_pred_df
```

```python
#checking for null values
flight_del_pred_df.isnull().sum()
```

```python
# remove last column with full null values
flight_del_pred_df.drop('Unnamed: 21', axis = 1, inplace = True)
flight_del_pred_df
```

```python
# Dropping all rows with null values
flight_del_pred_df.dropna(inplace = True)
flight_del_pred_df
```

```python
#checking for null values
flight_del_pred_df.isnull().sum()
```

```python
# checking for duplicates
flight_del_pred_df.duplicated().sum()
```

```python
# converting dummy variables(strings) to corresponding binary values
le = LabelEncoder()
dummy = ["OP_UNIQUE_CARRIER","OP_CARRIER","TAIL_NUM","ORIGIN","DEST","DEP_TIME","DEP_DEL15","DEP_TIME_BLK","ARR_TIME","ARR_DEL15","CANCELLED","DIVERTED","DISTANCE"]
for i in dummy:
    flight_del_pred_df[i] = le.fit_transform(flight_del_pred_df[i].astype(str))
flight_del_pred_df
    
```

```python
#checking for null values
flight_del_pred_df.isnull().sum()
```

```python
# Checking if any value in the column "CANCELLED" is greater than 0
is_greater_than_zero = (flight_del_pred_df['CANCELLED'] > 0).any()

if is_greater_than_zero:
    print("There are values greater than 0 in the column.")
else:
    print("There are no values greater than 0 in the column.")
```

```python
# Checking if any value in the column "DIVERTED" is greater than 0
is_greater_than_zero = (flight_del_pred_df['DIVERTED'] > 0).any()

if is_greater_than_zero:
    print("There are values greater than 0 in the column.")
else:
    print("There are no values greater than 0 in the column.")
```

```python
# remove "CANCELLED" and "DIVERTED" columns
flight_del_pred_df.drop(['CANCELLED', 'DIVERTED'], axis = 1, inplace = True)
flight_del_pred_df
```

```python
# Checking for correlation
flight_del_pred_df.corr()
```

```python
# Input dataset
x = flight_del_pred_df.drop(columns=flight_del_pred_df.columns[14])
x
```

```python
# output dataset 
y = flight_del_pred_df.iloc[:,14]
y
```

```python
# splitting to training and testing datasets
trainx, testx, trainy, testy = train_test_split(x, y, test_size = 0.1)
```

```python
# Random Forest model building
Random_Forest = RFC(n_jobs = 18, n_estimators = 100, oob_score = True, criterion = 'entropy')

# Train the model and predict the test data
modelrf = Random_Forest.fit(trainx,trainy).predict(testx)
```

```python
# Accuracy of Random Forest Model
accuracyrf = np.mean(modelrf == testy)
accuracyrf    
```
