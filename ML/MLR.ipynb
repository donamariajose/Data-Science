```python
# multiple linear rregression algorithm based data model
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# reading data
data=pd.read_csv('C:/Users/donad/OneDrive/Desktop/Internship/1_Machine Learning/4_Supervised_Learning/Regression/Datasets/StudentsPerformance.csv')
data
```

```python
# check for null value
data.isnull().sum()
```

```python
# check for dupplicates
data.duplicated().sum()
```

```python
# check data correlation
data.corr()
```

```python
# removing dummy variable
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
dummy=['gender','race/ethnicity','parental level of education','lunch','test preparation course']
for i in dummy:
    data[i]=le.fit_transform(data[i].astype(str))
data
```

```python
x=data.iloc[:,[0,1,2,3,4,6,7]]
x
```

```python
y=data.iloc[:,5]
y
```

```python
# splitting to train and test
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=True)

# model building
lr=LinearRegression()
lr.fit(xtrain,ytrain)
predlr=lr.predict(xtest)
predlr
```

```python
ytest
```

```python
# accuracy of model
score=r2_score(predlr,ytest)
score
```
