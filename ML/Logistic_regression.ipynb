```python
# Application logistic regression algorithm to build data model
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

# reading data
data=pd.read_csv('C:/Users/donad/OneDrive/Desktop/Internship/1_Machine Learning/4_Supervised_Learning/Classification/Datasets/diabetes (3).csv')
data
```

```python
# check for null value
data.isnull().sum()
```

```python
# check for dupplicates
data.duplicated().sum()
```

```python
# check data correlation
data.corr()
```

```python
# splitting to train and test
train,test=train_test_split(data,test_size=0.1)
train
```

```python
test
```

```python
trainx=train.iloc[:,:8]
trainy=train.iloc[:,8]
testx=test.iloc[:,:8]
testy=test.iloc[:,8]

# model building
lg=LogisticRegression()
predlg=lg.fit(trainx,trainy).predict(testx)

# accuracy of
acc=np.mean(predlg==testy)
acc
```
