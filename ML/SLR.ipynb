```python
# supervised learning
# simple linear regression algorithm based data model
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# reading data
data=pd.read_csv('C:/Users/donad/OneDrive/Desktop/Internship/1_Machine Learning/4_Supervised_Learning/Regression/Datasets/test.csv')
data
```

```python
# check for null value
data.isnull().sum()
```

```python
# check for dupplicates
data.duplicated().sum()
```

```python
# # check data correlation
data.corr()
```

```python
# splitting to train and test
x=data[['x']]
y=data[['y']]
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2)

# model building
slr=LinearRegression()
slr.fit(xtrain,ytrain)
predlr=slr.predict(xtest)
predlr
```

```python
ytest
```

```python
# printing mean squared error
print(mean_squared_error(predlr,ytest))
```

```python
# ploting graph
import matplotlib.pyplot as plt
plt.scatter(x=data['x'],y=data['y'])
plt.title('LinearRegression')
plt.grid()
plt.show()
```
