```python
import matplotlib.pyplot as plt
import numpy as np

graph_size = (10, 10)
fig,axs=plt.subplots(4,4,figsize=graph_size)

#plot1
x=np.array([10,20,30,40,50])
y=np.array([100,200,300,400,500])
plt.subplot(4,4,1)
plt.subplots_adjust(wspace=1,hspace=1)
plt.plot(x,y,marker='o',color='red')
plt.title('Rate of spread of Flue')
plt.xlabel('No. of days')
plt.ylabel('No. of people effected')

#plot2
x=np.array(['Jan','Feb','Mar','Apr','May'])
y=np.array([89,90,90,91,89])
plt.subplot(4,4,2)
plt.scatter(x,y,marker='D',color='blue')
plt.title('EUR to INR conversion')
plt.xlabel('x axis')
plt.ylabel('y axis')

#plot3
x=np.array([1,2,3,4,5])
y=np.array([30,36,29,40,41])
plt.subplot(4,4,3)
plt.bar(x,y,color='green')
plt.title('Students in UP class')
plt.xlabel('Class')
plt.ylabel('No.of students')

#plot4
x=np.array([1,2,3,4,5,6,7])
plt.subplot(4,4,4)
plt.boxplot(x)

#plot5
y1=np.array([3,8,1,10])
y2=np.array([6,2,7,11])
y3=np.array([7,3,12,20])
plt.subplot(4,4,5)
plt.plot(y1,color='red',marker='*')
plt.plot(y2,color='blue',marker='o')
plt.plot(y3,color='green',marker='D')
plt.title('Stock Market')
plt.xlabel('Date(January)')
plt.ylabel('rate of change')

#plot6
x=np.array([1,20,25,40,45])
y=np.array([100,200,300,400,500])
plt.subplot(4,4,6)
plt.scatter(x,y,marker='o',color='black')
plt.title('Sales of umbrella')
plt.xlabel('no.of days')
plt.ylabel('no.of umbrellas')

#plot7
x=np.array([1,2,3,4,5])
y=np.array([10,25,13,24,15])
plt.subplot(4,4,7)
plt.bar(x,y,color='blue')
plt.title('Eggs sold per carton')
plt.xlabel('type of carton')
plt.ylabel('No.of cartons')

#plot8
arr=np.array([35,25,25,15])
y=np.array(['apple','orange','kiwi','pinapple'])
color=['red','orange','green','yellow']
plt.subplot(4,4,8)
plt.pie(arr,labels=y,colors=color)
plt.title('Fruits bought per month')

#plot9
x=np.array([10,20,30,40,50])
y=np.array([50,20,100,40,77])
plt.subplot(4,4,9)
plt.plot(x,y,marker='o',color='blue')
plt.title('Tickets sold for show')
plt.xlabel('Time in inutes')
plt.ylabel('No.of tickets')

#plot10
x=np.array([10,20,25,40,50])
y=np.array([100,500,300,400,200])
plt.subplot(4,4,10)
plt.plot(x,y,marker='*',color='green')
plt.title('Visitors in RK park')
plt.xlabel('Days')
plt.ylabel('No.of people')

#plot11
x=np.array(['Mon','Tue','Wed','Thu','Fri'])
y1=np.array([3,2,1,4,1])
y2=np.array([4,2,3,4,3])
y3=np.array([1,3,2,2,4])
plt.subplot(4,4,11)
plt.plot(x,y1,color='orange',marker='*')
plt.plot(x,y2,color='red',marker='o')
plt.plot(x,y3,color='green',marker='D')
plt.title('Film shows per day')
plt.xlabel('days')
plt.ylabel('y axis label')

#plot12
x=np.array([9,6,3,9,5,6,1])
plt.subplot(4,4,12)
plt.boxplot(x)

#plot13
x=np.array([1,2,3,4,5])
y=np.array([60,31,54,10,11])
plt.subplot(4,4,13)
plt.bar(x,y,color='red')
plt.title('Fish cans sold per day')
plt.xlabel('days')
plt.ylabel('y axis label')

#plot14
x=np.array([1,2,3,4,5])
y=np.array([100,200,300,400,500])
plt.subplot(4,4,14)
plt.scatter(x,y,marker='D',color='orange')
plt.title('Balls in boxes')
plt.xlabel('Box labels')
plt.ylabel('no.of balls')

#plot15
x=np.array(['A','B','C','D','E'])
y=np.array([5,2.5,3.6,4.8,4.5])
plt.subplot(4,4,15)
plt.bar(x,y,color='blue')
plt.title('CC Ratings for Films')
plt.xlabel('days')
plt.ylabel('y axis label')

#plot16
arr=np.array([35,25,25,15])
y=np.array(['food','education','rent','clothing'])
color=['green','blue','grey','yellow']
plt.subplot(4,4,16)
plt.pie(arr,labels=y,colors=color)
plt.title('Money spent per items')
```
