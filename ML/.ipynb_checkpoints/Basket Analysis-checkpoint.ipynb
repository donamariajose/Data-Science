```python
# association  data model based on Apriori algorithm
import pandas as pd
import numpy as np
from apyori import apriori

# reading data
ds=pd.read_csv('C:/Users/donad/OneDrive/Desktop/Internship/1_Machine Learning/5_Unsupervised_Learning/3_Association/dataset/Groceries_dataset.csv')
ds
```

```python
# data[['col1', 'col2', 'col3','col4', 'col5', 'col6']] = data['itemDescription'].str.split(',', expand=True)
split_data = data['itemDescription'].str.split(',', expand=True)
split_data
# # Rename the columns
# split_data.columns = ['col1', 'col2', 'col3', 'col4', 'col5', 'col6','col7', 'col8', 'col9', 'col10', 'col11']
# split_data.columns
# # # Concatenate the split columns with the original DataFrame
# data = pd.concat([data, split_data], axis=1)
# data
```

```python
split_data.isnull().sum()
```

```python
split_data.fillna(value=np.nan, inplace=True)
split_data
```

```python
records=[]
for i in range(0,14963):
  records.append([str(split_data.values[i,j])for j in range(0,11)])

records[0]
# records = []
# for i in range(0, 7501):
#     records.append([str(split_data.values[i,j]) for j in range(0, 10)])

# records[0]
```

```python
# item set association
association_rules = apriori(records, min_support=0.00030, min_confidence=0.05)
association_results = list(association_rules)

print("There are {} Relation derived.".format(len(association_results)))
```

```python
# printing associations
for i in range(0, len(association_results)):
  print(association_results[i][0])
```

```python
# association in dataframe
def inspect(results):
  print(len(results))
  lhs=[tuple(result[2][0])[0] for result in results]
  rhs=[tuple(result[2][0][1])[0] for result in results]
  supports=[result[1] for result in results]
  confidences=[result[2][0][2] for result in results]
  lifts=[result[2][0][3] for result in results]
  return list (zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(association_results), columns = ["Left hand side", "Right hand side", "Support", "Confidence", "Lift"])
resultsinDataFrame
```

```python
# printing first 10 rows
resultsinDataFrame.nlargest(n = 10,columns = "Lift")
```
