```python
# kmeans algorithm based data model
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cluster import KMeans

# reading data
ds=pd.read_csv('C:/Users/donad/OneDrive/Desktop/Internship/1_Machine Learning/5_Unsupervised_Learning/1_clusters/datasets/Mall_Customers.csv')
ds
```

```python
# removing dummy variable
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['Gender']=le.fit_transform(ds['Gender'].astype(str))
ds
```

```python
x=ds.iloc[:,:4]
x
```

```python
wcss=[]
for i in range(1,15):
    kmeans=KMeans(n_clusters=i,init='k-means++',random_state=1)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
wcss
```

```python
# ploting graph
plt.plot(range(1,15),wcss)
plt.title('elbow methods')
plt.xlabel('number of cluster')
plt.ylabel('wcss')
plt.show()
```

```python
# clustering
kmeans=KMeans(n_clusters=4,init='k-means++',random_state=1)
y=kmeans.fit_predict(x)

# changing array to column
s=pd.Series(y)
s
```

```python
# adding new column to dataset
ds['clust']=s
ds
```

```python
ds.iloc[:,[5,0,1,2,3,4]]
```

```python
# grouping data based on cluster
ds.groupby(ds.clust).mean()
```
