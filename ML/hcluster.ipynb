```python
# hierarchical cluster algorithm based data model
import pandas as pd
import matplotlib.pylab as plt
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch

# reading data
ds=pd.read_csv('C:/Users/donad/OneDrive/Desktop/Internship/1_Machine Learning/5_Unsupervised_Learning/1_clusters/datasets/Mall_Customers.csv')
ds
```

```python
# removing dummy variable
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
ds['Gender']=le.fit_transform(ds['Gender'].astype(str))
ds
```

```python
# normalize-mean-0, sd-1
def nor_fun(i):
    x=(i-i.mean())/(i.std())
    return x
norm_fun=nor_fun(ds.iloc[:,1:])
norm_fun
```

```python
# dendrogram
z=linkage(norm_fun,method='complete',metric='euclidean')

# ploting graph
plt.figure(figsize=(50,10));plt.title('Dendrogram');plt.xlabel=('index');plt.ylabel=('Distance')
sch.dendrogram(z,leaf_rotation=0,leaf_font_size=8)
plt.show()
```

```python
# clustering
from sklearn.cluster import AgglomerativeClustering
a=AgglomerativeClustering(n_clusters=4,linkage='complete',affinity='euclidean').fit(norm_fun)
a.labels_
```

```python
# changing array to column
labels=pd.Series(a.labels_)
labels
```

```python
# adding new column to dataset
ds['cluster']=labels
ds
```

```python
# grouping data based on cluster
ds.groupby(ds.cluster).mean()
```
