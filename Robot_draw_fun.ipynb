```python
# drawing a robot using opencv draw functions
import numpy as np
import cv2
from matplotlib import pyplot as plt

# creating white background
bg = np.ones((500,500,3), dtype=np.uint8)*255
cv2.imshow('backgroung',bg)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
# ploting bg as graph
bg = np.ones((500,500,3), dtype=np.uint8)*255
plt.xticks(),plt.yticks()
plt.title('Background')
plt.imshow(bg)
```

```python
# drawing robot
bg = np.ones((500,500,3), dtype=np.uint8)*255

# Head
cv2.circle(bg,(250,150),50,(127,127,127),-4)
# eyes
cv2.circle(bg,(225,140),10,(255,255,255),-4)
cv2.circle(bg,(275,140),10,(255,255,255),-4)
cv2.circle(bg,(275,142),5,(0,0,0),-4)
cv2.circle(bg,(225,142),5,(0,0,0),-4)
# eyebrows
cv2.line(bg,(218,125),(232,125),(0,0,0),3)
cv2.line(bg,(268,125),(282,125),(0,0,0),3)
# nose
cv2.line(bg,(248,165),(252,165),(77,77,77),4)
# mouth
cv2.line(bg,(240,175),(260,175),(0,0,0),2)

# body
cv2.rectangle(bg,(200,200),(300,375),(127,127,127),-4)
# hands
cv2.rectangle(bg,(180,210),(200,330),(77,77,77),-4)
cv2.rectangle(bg,(300,210),(320,330),(77,77,77),-4)
# fingers
cv2.line(bg,(180,330),(180,340),(77,77,77),4)
cv2.line(bg,(186,330),(186,340),(77,77,77),4)
cv2.line(bg,(192,330),(192,340),(77,77,77),4)
cv2.line(bg,(198,330),(198,340),(77,77,77),4)
cv2.line(bg,(300,330),(300,340),(77,77,77),4)
cv2.line(bg,(306,330),(306,340),(77,77,77),4)
cv2.line(bg,(312,330),(312,340),(77,77,77),4)
cv2.line(bg,(318,330),(318,340),(77,77,77),4) 
# legs
cv2.rectangle(bg,(210,375),(245,475),(77,77,77),-4)
cv2.rectangle(bg,(255,375),(290,475),(77,77,77),-4)

# ploting
plt.xticks(),plt.yticks()
plt.title('ROBOT')
plt.imshow(bg)
```
